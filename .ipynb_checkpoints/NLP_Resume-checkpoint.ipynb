```python
# importing required modules
import PyPDF2
import nltk
import os
from os import walk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
import helper

 
r_dir = 'resumes/'                                                   # contains all file types of resumes
r_text_dir = 'resumes_text/'                                         # stores the text version of resume
```

```python
f = helper.get_all_files(r_dir)                                      # get all files in resumes directory

for file_ in f:
    filename, file_extension = os.path.splitext(file_)                          # get file name, extension
    
    if file_extension.lower() == '.pdf':                        # if the file type is pdf, save the text file version
        file_text = helper.get_pdf_corpus(r_dir + file_)                  # read pdf text  
        helper.save_text_file( r_text_dir + filename + '.txt', file_text)                 # save as text file
```
